```python
import requests
from selenium import webdriver
import pandas as pd

PATH = 'chromedriver.exe'
driver = webdriver.Chrome(PATH)
liste = []
i = 1000000
for j in range(15100): #max 15100
    i= i+1
    req = 'https://procinema.ch/de/statistics/filmdb/', str(i), '.html'
    new_req = ''.join(req)
    request = requests.get(new_req)
    if request.status_code == 200:
        print(new_req,'Web site exists')
        driver.get(new_req)
        
        #Hauptinformation zum Film
        try:
            ls1 = []
            all_faditemlbl = driver.find_elements_by_xpath("//div[@class='faditemlbl']")
            for faditemlbl in all_faditemlbl:
                ls1.append(faditemlbl.text)
            df1 = pd.DataFrame(data=ls1)

            ls2 = []
            all_faditemcont = driver.find_elements_by_xpath("//div[@class='faditemcont']")
            for faditemcont in all_faditemcont:
                ls2.append(faditemcont.text)
            df2 = pd.DataFrame(data=ls2)
        except:
            raise ValueError('Hauptinformationen zum Film konnten nicht geholt werden')


        #DETAILS ZUM FILM
        try:
            ls3 = []
            all_extras = driver.find_elements_by_class_name('extras')
            for extras in all_extras:
                ls3.append(extras.text)  

            str3 = ''.join(ls3)
            str3s = str3.split("\n")
            ls3s = str3s[4:]
            ls3s1 = []
            ls3s2 = []
            if len(ls3s[-1]) == 1:
                ls3s = ls3s[:-1]
            for m in range(len(ls3s)):
                ls3s[m] = ls3s[m].replace("  ", "")
                split_1 = ls3s[m].split(": ")
                ls3s1.append(split_1[0])
                ls3s2.append(split_1[1])
            df3s1 = pd.DataFrame(data=ls3s1)
            df3s2 = pd.DataFrame(data=ls3s2)
            df_row2 = pd.concat([df3s1, df3s2],axis=1)
        except:
            raise ValueError('Details zum Film konnten nicht geholt werden')

        
        #Bild mit text abspeichern
        try:
            l=driver.find_element_by_css_selector("h1")
            test = 'Bilder/', l.text, '.png'
            new_liste = ''.join(test)
            driver.save_screenshot(new_liste)
        except:
            raise ValueError('Titel konnte nicht geholt werden und dadurch wurde das Bild nicht gespeichert')

        try:
            df_row1 = pd.concat([df1, df2],axis=1)
            df_row = pd.concat([df_row1, df_row2],axis=0)
            df_row_T = df_row.T
            df_row_T.columns = df_row_T.iloc[0]
            df_row_T.reset_index()
            df_row_T = df_row_T[1:]
            cols = []
            count = 1
            for column in df_row_T.columns:
                if column == 'Schweiz':
                    cols.append(f'Schweiz_{count}')
                    count+=1
                    continue
                cols.append(column)
            df_row_T.columns = cols
            df_row_T = df_row_T.drop(['Deutschschweiz','Suisse romande','Tessin','Link'], axis=1, errors='ignore')
            liste.append(df_row_T)
        except:
            raise ValueError('Fehler im Dataframe. Dataframe konnte nicht zusammengefhrt werden.')
                
    else:
        print(new_req,'Web site does not exist') 
print("done")
driver.quit()

df = pd.concat(liste, ignore_index=True, sort=False)
df_f = df[["Original", "Schweiz_2","Regie","Produzent","Schauspieler","Produktionsjahr", "Produktionsland", "Genre", "Originalsprache","Filmlnge"]]
df.to_csv("data/data.csv", index=False)
```
