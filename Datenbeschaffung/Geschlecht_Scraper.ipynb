```python
import pandas as pd
df = pd.read_csv("../data/data.csv")
df_split = df.assign(Schauspieler=df['Schauspieler'].str.split('\n')).explode('Schauspieler')
df_split = df_split.reset_index()
pd.set_option('display.max_rows', None)
df_split.to_csv("../data/data_split.csv", index=False)
```

```python
import time

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

PATH = 'chromedriver.exe'
driver = webdriver.Chrome(PATH)
namen_liste = []
sex_list =[]
for i in range(len(df_split)):
    df_name = pd.DataFrame({'namen_liste':namen_liste,'sex_list':sex_list})
    if type(df_split.Schauspieler[i]) is str:
        try:
            splited = df_split.Schauspieler[i].split(' ')
            mask = df_name[(df_name['namen_liste'] == splited[0])]
            if splited[0] in df_name.namen_liste.unique():
                print(splited[0], 'in own database')
                sex_list.append(mask.iloc[0][1])
                namen_liste.append(splited[0])
            else:    
                driver.get("https://www.babycenter.com/baby-names")
                current_url = driver.current_url
                input_element = driver.find_element_by_name("name")
                input_element.send_keys(splited[0])
                input_element.submit()
                WebDriverWait(driver, 15).until(EC.url_changes(current_url))
                l=driver.find_element_by_css_selector("h1")
                test = l.text
                if "Girl" in test: 
                    sex_list.append('female')
                    namen_liste.append(splited[0])
                    print(splited[0], 'is female')
                elif "Boy" in test :
                    sex_list.append('male')
                    namen_liste.append(splited[0])
                    print(splited[0], 'is male')
                else:
                    sex_list.append('-')
                    namen_liste.append(splited[0])
                    print(splited[0], 'not in this Website')
                    time.sleep(4)
        except:
            sex_list.append('-')
            namen_liste.append('-')
            print(splited[0], 'not in this Website')
    else:
        sex_list.append('-')
        namen_liste.append('-')
        print(df_split.Schauspieler[i], 'missing actor')
driver.quit()
df_split['sex'] = sex_list
df_split.to_csv("data/data_complete.csv", index=False)
```
